To access a specific value in a column, use tblname[tblname.colname  == "val"]

```python
prop = props[props.prop_name == "PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE."]
```

```python
props.prop_name.value_counts()
```

Use reset_index() to convert series to a data frame

```python
props.prop_name.value_counts().reset_index()
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.head()
```

To merge databases, identify a foreign key. By default, it is "inner join"

```python
contribs.info()
```

```python
prop.info()
```

```python
merged = pd.merge(prop, contribs, on = "calaccess_committee_id" )
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / merged.amount.sum()
```

To sort, type tblname.sort_values("Colname")

```python
merged.sort_values("amount", ascending= False)
```

To group, use tblname.groupby("colname").colname.method()

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending = False)
```

To merge by multiple cols, use tblname.groupby["col1","col2"].colname.method()

```python
merged.groupby(["contributor_lastname", "contributor_firstname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending = False)
```

```python
top_supporters = merged.groupby(["contributor_lastname", "contributor_firstname"]).amount.sum().reset_index().sort_values("amount", ascending = False)
```

To graph data, use plot.bar() method

```python
chart = top_supporters.amount.head(10).plot.bar()
chart.set_xticklabels(top_supporters.contributor_fullname)
```

```python
chart2 = top_supporters.amount.head(10).plot.barh()
chart2.set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters['contributor_fullname'] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
top_supporters.head(5).to_csv("top_supporters.csv")
```
